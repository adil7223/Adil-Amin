# 16-Splitting a String into Multiple Columns

```python
#import libraries
import pandas as pd
df = pd.DataFrame({'name':['muhammad usman','waleed hayat','noman akhter','malik salman','sohail marwat'],
                  'location':['lahore, pakistan','karachi, pakistan','gujranwala, pakistan','peshawar, pakistan','bannu, pakistan']})
df
```

```python
#splitting a column into two columns
df.name.str.split(' ', expand=True)
```

```python
df
```

```python
#adding those splits into new columns
df[['first_named','last_name']] = df.name.str.split(' ', expand=True)
df.location.str.split(' ', expand=True)
df[['city','country']] = df.location.str.split(' ', expand=True)
df
```

```python
df = df[['first_named','last_name','city','country']]
df
```
